```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
L = [[1, 1]]
N = 500
for i in range(N):
    deltaX1 = lambda x1, x2: 1 - x1 - 1.4*x1**2 + x2
    deltaX2 = lambda x1, x2: 0.3*x1 - x2
    x1_last, x2_last = L[-1][0], L[-1][1]
    x1_next = x1_last + deltaX1(x1_last, x2_last)
    x2_next = x2_last + deltaX2(x1_last, x2_last)
    L.append([x1_next, x2_next])
```

```python
L = np.array(L)
plt.scatter(L[:,0], L[:,1])
```

```python
plt.plot(L[:,0], L[:,1])
```

```python
L = [[25, 100]]
N = 5000
for i in range(N):
    deltaX1 = lambda x1, x2: -0.01*x1 + 0.0001*x1*x2
    deltaX2 = lambda x1, x2: 0.0005*x2 - 0.0001*x1*x2
    x1_last, x2_last = L[-1][0], L[-1][1]
    x1_next = x1_last + deltaX1(x1_last, x2_last)
    x2_next = x2_last + deltaX2(x1_last, x2_last)
    L.append([x1_next, x2_next])
```

```python
L = np.array(L)
plt.scatter(L[:,0], L[:,1])
```

```python
# Plot Predator and Prey Separately against time
pred = [a for a, b in L]
prey = [b for a, b in L]
t = np.linspace(0, 5001, 5001)
plt.scatter(t, pred, color= 'red', label= 'predator')
plt.scatter(t, prey, color= 'blue', label= 'prey')
plt.legend(loc= 'upper left')
```
