#  RLC Circuit Simulation

**Dynamical System of the RLC Circuit**

$\cfrac{dx_1}{dt} = x_1 -x_1^3 - x_2$

$\cfrac{dx_2}{dt} = 3x_1$

$x_1 = \text{current}$

$x_2 = \text{voltage}$

$x_1(0) = -1$

$x_2(0) = -1.5$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
init = [[-1, -1.5]]
N = 100
delT = 0.1
def f1(x1, x2): return x1 - x1**3 - x2
def f2(x1, x2): return 3*x1 
for i in range(N):
    x1_Old, x2_Old = init[-1][0], init[-1][1]
    x1_New = x1_Old + delT*f1(x1_Old, x2_Old)
    x2_New = x2_Old + delT*f2(x1_Old, x2_Old)
    init.append([x1_New, x2_New])
    
init = np.array(init)
```

```python
init_ = [[0.3, 0.1]]
N_ = 100
delT_ = 0.1

for i in range(N_):
    x1_Old, x2_Old = init_[-1][0], init_[-1][1]
    x1_New = x1_Old + delT_*f1(x1_Old, x2_Old)
    x2_New = x2_Old + delT_*f2(x1_Old, x2_Old)
    init_.append([x1_New, x2_New])
    
init_ = np.array(init_)
```

```python
fig, ax = plt.subplots()
ax.plot(init[:, 0], init[:, 1], c= 'r', label= 'init = [[-1, -1.5]]')
ax.plot(init_[:, 0], init_[:, 1], c= 'b', label= 'init_ = [[0.3, 0.1]]')
ax.set(xlabel= 'current', ylabel= 'voltage', title= 'RLC Circuit Simulation')
ax.legend(loc= 'upper right')
ax.grid()
```
