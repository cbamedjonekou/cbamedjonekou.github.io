## Chaos Simulation

**Assumptions**

$x_1; x_2 \text{ Are the State Variables}$

**Given Continuous Dynamical System**

$\cfrac{dx_1}{dt} = (a-1)x_1 + ax_1^2$

$\cfrac{dx_2}{dt} = x_1 - x_2$


**Give Discrete Dynamical System**

$\Delta x_1 = (a-1)x_1 + ax_1^2$

$\Delta x_2 = x_1 - x_2$

**Objective**

1. Find equilibrium Points
    * Prove that equilibrium pts for the continuous dynamical system are stable at $x_1 = x_2 = (a-1) = a$ for any $a > 1$
    * Prove that equilibrium pts for the discrete dynamical system are stable at $x_1 = x_2 = \frac{(a-1)}{a}$ for any $a > 1$
    
    
2. Run a simulation w/ various the values: $a = [1.5, 2.0, 2.5, 3.0, 3.5, 4.0]$
    * You can choose arbitrary stating points for each run

### Packages

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sympy.solvers import solve
import sympy as sp
from School_Toolkit import Mathematical_Modeling as mm
from scipy.linalg import eig
import math as m
```

### Setting Up the Dynamical Systems

```python
# Print formating
sp.init_printing()

# Various values for 'a'
initl = [1.5, 2.0, 2.5, 3.0, 3.5, 4.0]

# Continuous Dynamical System
x1, x2, a = sp.symbols('x1 x2 a') 
dx1, dx2 = (a-1)*x1 - a*x1**2, x1 - x2


# Discrete Dynamical System
delta_x1, delta_x2 = (a-1)*x1 - a*x1**2, x1 - x2
G1, G2 = x1 + delta_x1, x2 + delta_x2
```

### Finding the Equilibrium Points

```python
# Continuous
sol = solve([dx1, dx2], x1, x2)
sol
```

```python
# Discrete
sol = solve([delta_x1, delta_x2], x1, x2)
sol
```

### Classify the Equilibrium Points

**Continuous Jacobian**

```python
sp.Matrix([dx1, dx2]).jacobian([x1, x2])
```

```python
points = [[sol[1][0], sol[1][1]]]
sp.Matrix(mm.poorManJacobian2(dx1, dx2, x1, x2, points, False))
```

**Discrete Jacobian**

```python
sp.Matrix([G1, G2]).jacobian([x1, x2])
```

```python
points = [[sol[1][0], sol[1][1]]]
sp.Matrix(mm.poorManJacobian2(G1, G2, x1, x2, points, False))
```

**Continuous looks stable, while the Discrete looks unstable**

### Simulations

```python
def f1(x1, x2, a):
    return (a-1)*x1 - a*x1**2

def f2(x1, x2, a):
    return x1 - x2

def continuous(a, N, L):
    for i in range (N):
        x1_Old=L[-1][0]
        x2_Old=L[-1][1]
        x1_New=x1_Old + f1(x1_Old, x2_Old, a)
        x2_New=x2_Old + f2(x1_Old, x2_Old, a)
        L.append([x1_New,x2_New])
    return np.array(L)

def discrete(a, N, L, delT= 1):
    for i in range (N):
        x1_Old=L[-1][0]
        x2_Old=L[-1][1]
        x1_New=x1_Old + delT* f1(x1_Old, x2_Old, a)
        x2_New=x2_Old + delT* f2(x1_Old, x2_Old, a)
        L.append([x1_New,x2_New])
    return np.array(L)
```

### Continuous

```python
ini = [[.2, .3]]
N = 100
s10 = continuous(initl[0], N, ini)
t = np.linspace(0, s10.shape[0], s10.shape[0])
plt.plot(t, s10[:,0], label= 'x1', c = 'r')
plt.plot(t, s10[:,1], label= 'x2', c = 'b')
plt.xlim(0, 20)
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('a = 1.5')
plt.legend(loc= 'lower right')
```
