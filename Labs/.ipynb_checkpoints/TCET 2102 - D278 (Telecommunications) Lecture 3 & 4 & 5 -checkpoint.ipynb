** Filters: Any combination of passive (R, L, and C) and/or active (transistors or operational amplifiers) elements designed to select or reject a band of frequencies is called a filter.**

*  In communication systems, filters are employed to pass those frequencies containing the desired information and to reject the remaining frequencies.

**Filters are employed to filter out any unwanted frequencies, com- monly called noise, due to the nonlinear characteristics of some elec- tronic devices or signals picked up from the surrounding medium. In general, there are two classifications of filters:**

* Passive Filters - contains only passive elements (e.g: resistors, capacitors, inductors) **"Passive filters are those filters composed of series or parallel combinations of R, L, and C elements."**

    * Low Pass Filters
    * High Pass Filters
    * Passive Band Filters
    * Stop Band Filters


* Active Filters - contain active elements (i.e.: diodes, Transistors, O.A) in addition to passive elements. **"Active Filters are filters that employ active devices such as transistors and operational amplifiers in combination with R, L, and C elements."** 

### Low Pass Filter

```python
import math as m
def reactanceCapacitor(frequency, Capacitance):
    """Calculates the reactance of a capacitor. Takes in 2 arguments: Frequency, Capacitance"""
    if frequency == 0:
        return m.inf
    else:
        return 1/(2*m.pi*frequency*Capacitance)
```

** The R-C filter is incredibly simple in design, and can be used as a low-pass or a high-pass filter. **

* If the output is taken off the capacitor, it will respond as a low-pass filter.
* The circuit should behave in a manner that will result in a high-level output for low frequencies and a declining level for frequencies above the critical value.

** Deriving characteristic equations for Low-Pass Filters **

* First we examine the network at the frequency extremes: no frequency ($f = 0 \space Hz$), and very high frequencies. We do this to test the response of the circuit. In doing so, we get the following:

    * **At $f = 0 \space Hz$ you get the following (which represents an open circuit):**
    
    $$X_C = \frac{1}{\omega C} = \frac{1}{2\pi f C} = \infty \Omega$$  
    <p></p>
    * **At very high frequencies $(\lim_{f\to\infty} \omega(f) \cong 0 \space Hz$) you get the following (which represents a closed circuit):**
    
    $$X_C = \frac{1}{\omega C} = \frac{1}{2\pi f C} \cong 0 \Omega$$  
    <p></p>
    * ***For filters, a normalized plot is used. Normalization is a process where a quantity (e.g.:voltage, current, or impedance) is divided by a quantity of the same unit of measurement to establish a dimensionless level of a specific value (or range). A normalized plot in the filter domain can be obtained by dividing the plotted quantity such as $V_O$ with the applied voltage $V_I$ for the frequency range of interest.***
    <p></p>
    * **Since the maximum value of $V_O$ for the low-pass filter of is $V_I$, each level of $V_O$ in the figure below is divided by the level of $V_I$. The result is the plot of $A_V = \frac{V_O}{V_I}$ of the figure below. Note that the maximum value is 1 and the cutoff frequency is defined at the 0.707 level.***
