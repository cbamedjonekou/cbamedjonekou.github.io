** Filters: Any combination of passive (R, L, and C) and/or active (transistors or operational amplifiers) elements designed to select or reject a band of frequencies is called a filter.**

*  In communication systems, filters are employed to pass those frequencies containing the desired information and to reject the remaining frequencies.

**Filters are employed to filter out any unwanted frequencies, commonly called noise, due to the nonlinear characteristics of some elec- tronic devices or signals picked up from the surrounding medium. In general, there are two classifications of filters:**

* Passive Filters - contains only passive elements (e.g: resistors, capacitors, inductors) **"Passive filters are those filters composed of series or parallel combinations of R, L, and C elements."**

    * Low Pass Filters
    * High Pass Filters
    * Passive Band Filters
    * Stop Band Filters


* Active Filters - contain active elements (i.e.: diodes, Transistors, Operational Amplifiers) in addition to passive elements. **"Active Filters are filters that employ active devices such as transistors and operational amplifiers in combination with R, L, and C elements."**

### Deriving characteristic equations for Low-Pass Filters 

*** First we examine the network at the frequency extremes: no frequency ($f = 0 \space Hz$), and very high frequencies. We do this to test the response of the circuit. In doing so, we get the following***:

* **At $f = 0 \space Hz$ you get the following (which represents an open circuit, the capacitor gets replaced w/ an open circuit). For open circuits $V_O = V_I$:**
    
$$X_C = \frac{1}{\omega C} = \frac{1}{2\pi f C} = \infty \Omega$$

$$X_C = \frac{1}{\omega C} = \frac{1}{2\pi f C} \cong 0 \Omega$$

* ***For filters, we use a normalized plot more often than the $V_O \space vs. \space frequency$ to show the characteristics of filters.***

    * **Normalization is a process where a quantity (e.g.:voltage, current, or impedance) is divided by a quantity of the same unit of measurement to establish a dimensionless level of a specific value (or range).*** A normalized plot in the filter domain can be obtained by dividing the plotted quantity such as $V_O$ with the applied voltage $V_I$ (***Voltage Gain $A_V = \frac{V_O}{V_I}$***) for the frequency range of interest.

* **Since the maximum value of $V_O$ for the low-pass filter of is $V_I$, each level of $V_O$ in the figure below is divided by the level of $V_I$. The result is the plot of $A_V = \frac{V_O}{V_I}$ of the figure below. Note that the maximum value is 1 (Given that $A_V = \frac{V_O}{V_I}; \space and \space V_O = V_I$) and the cutoff frequency is defined at the 0.707 level.***
    

### At any frequency in between the extremes, the output voltage $V_O$ of the figure below can be determined using the voltage divider rule: **



$$V_O = \frac{X_C \angle{-90^{\circ} V_I}}{R-jX_C}$$



* **We can define this relationship in terms of Voltage Gain, $A_V$, as it would be in terms of our *normalized plot* **


$$A_V = \frac{V_O}{ V_I} = \frac{X_C \angle{-90^{\circ}}}{R-jX_C}$$

* **Which can be expanded to the following (by converting the denominator from rectangular to polar form.**

$$A_V = \frac{V_O}{ V_I} = \frac{X_C \angle{-90^{\circ}}}{\sqrt{R^2+X_C^2} \angle{-\tan^{-1}\frac{X_C}{R}}} = \frac{X_C}{\sqrt{R^2+X_C^2}} \angle{-90 + \tan^{-1}\frac{X_C}{R}}$$

* **Further Simplifying the magnitude, we get the following: **

$$A_V = \frac{V_O}{ V_I} = \frac{X_C}{\sqrt{R^2+X_C^2}} = \frac{1}{\sqrt{\left(\frac{R}{X_C}\right)^2 + 1}}$$ 

* **Simplifying the phase angle, we get the following:**

$$ \angle{-90 + \tan^{-1}\frac{X_C}{R}} = \angle{- \tan^{-1}\frac{R}{X_C}}$$

* **We can also define this relationship in terms of frequency by defining $R$ and $X_C$ in terms of frequency, shown below:**

$$A_V = \frac{V_O}{ V_I} =  \frac{1}{\sqrt{\left(\frac{R}{X_C}\right)^2 + 1}} \angle{- \tan^{-1}\frac{R}{X_C}}$$  

* **For $R = \frac{1}{2\pi f C}$ and $X_C = \frac{1}{2\pi f_C C}$**
$$=  \frac{1}{\sqrt{\left(\frac{ \frac{1}{2\pi f C}}{ \frac{1}{2\pi f_C C}}\right)^2 + 1}} \angle{- \tan^{-1}\frac{\frac{1}{2\pi f C}}{ \frac{1}{2\pi f_C C}}}$$  

$$=  \frac{1}{\sqrt{\left(\frac{1}{2\pi f C} * \frac{2\pi f_C C}{1}\right)^2 + 1}} \angle{- \tan^{-1}\frac{1}{2\pi f C} * \frac{2\pi f_C C}{1}}$$ 

$$=  \frac{1}{\sqrt{\left(\frac{f_C}{f}\right)^2 + 1}} \angle{- \tan^{-1}\frac{f_C}{f}}$$ 

$$\textbf{Note that } f_C = \frac{1}{2\pi RC}$$

### For the special frequency at which $X_C = R$ or $f = f_C$ , the magnitude becomes

$$A_V = \frac{V_O}{ V_I} = \frac{1}{\sqrt{\left(\frac{R}{X_C}\right)^2 + 1}} = \frac{1}{\sqrt{\left(\frac{f_C}{f}\right)^2 + 1}} = \frac{1}{\sqrt{2}} \cong 0.707$$

### The impact of the equation $f_C = \frac{1}{2\pi RC}$, extends beyond their relative simplicity. 

* **For any low-pass filter, the application of any frequency less than $f_C$ will result in an output voltage $V_O$ that is at least $70.7\%$ of the maximum.** 

$$\text{For } \space f < f_C, \space V_O \geq 0.707 V_I$$   

* **For any frequency above $f_C$, the output is less than $70.7\%$ of the applied signal.**

$$\text{For } \space f > f_C, \space V_O < 0.707 V_I$$  

* **And at $f_C$, the following is a true observed characteristic.**

$$\mathbf{V_O} \space lags \space \mathbf{V_I} \space by \space 45^{\circ}$$

**A number of methods are used to establish the pass-band characteristic of the Figure above. The one method being discussed here uses both a low-pass and a high-pass filter in cascade, as shown in the Figure below.**

```python
i(filename='/Users/Chris/Desktop/Images/TCET11.png', width= 400, height= 400)
```

**Having two filters, a low-pass and high-pass cascaded, requires you to realize the following characteristics:**

* **You have two frequencies that pass through your pass-band filter due to the nature of it's two components, the low-pass and high-pass filters**.

    * The low-pass filters, as the name suggest, allows **smaller/lower frequencies to pass** through as they have **outputs (voltage) which are greater than $70.7\%$ of their inputs (voltage),** ($\mathbf{f < f_C; \space V_O > 0.707 V_I \space}$). They block anything higher than the critical frequency ($\mathbf{f = f_C; \space V_O = 0.707 V_I \space}$) from going through, ($\mathbf{f > f_C; \space V_O < 0.707 V_I \space}$).
    <p></p>
    * The high-pass filters, as the name suggest, allows **larger/higher frequencies to pass** through as they have **outputs (voltage) which are greater than $70.7\%$ of their inputs (voltage),** ($\mathbf{f > f_C; \space V_O > 0.707 V_I \space}$). They block anything lower than the critical frequency ($\mathbf{f = f_C; \space V_O = 0.707 V_I \space}$) from going through, ($\mathbf{f < f_C; \space V_O < 0.707 V_I \space}$).
    
    
* **You have two critical/cuttoff frequencies; the area formed between them is called the Bandwidth**.

    *  If the critical frequencies of each filter gets closer and closer together, then there's an increase to the quality factor of the response curve causing the peak values within the pass-band to continue to drop. 


* **You have a frequency, $f_O$, near the center of the pass-band that will pass through both filters with very little degeneration**.

    * For cases where $V_{O(max)} \neq V_{I(max)}$ the bandwidth is defined at $0.707$ of the resulting $V_{O(max)}$.

$$f = f_O; BW = 0.707 V_{O(max)} $$

    * ** However, $V_O$ will not be equal to $V_I$ in the pass-band, but a frequency range in which $V_O$ will be equal to or greater than $0.707*V_{max}$ can be defined.**
