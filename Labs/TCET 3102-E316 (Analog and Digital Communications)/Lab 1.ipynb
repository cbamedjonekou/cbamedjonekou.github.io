### Modules (Packages)

```python
# These are the packages I'll need to solve this problem
import math as m
import numpy as np
from matplotlib import pyplot as plt
```

### Variables for time domain plot

```python
# These variables 
start, stop, n1, n2, samplingfreq = 0, 100, 9, 91, 100000 
t = np.arange(start, stop, .001) / samplingfreq
A = 2
fundamental = A/2
signalfreq = 1000
omega = 2 * np.pi * signalfreq
harmonics1 = sum([((2*A)/(np.pi*(2*p+1))) * np.sin((2*p+1) * omega * t) for p in range(n1+1)])
harmonics2 = sum([((2*A)/(np.pi*(2*p+1))) * np.sin((2*p+1) * omega * t) for p in range(n2+1)])
ffs1, ffs2 = lambda n: (fundamental + harmonics1), lambda w: (fundamental + harmonics2)
```

### Plot of the fourier series in the time domain

```python
# Creates the figure and subplot 1
fig, ax1 = plt.figure(figsize= (10,10)), plt.subplot(2, 1, 1)
ax1.plot(t, ffs1(n1))

# Creates subplot 2
ax2 = plt.subplot(2, 1, 2)
ax2.plot(t, ffs2(n2), color= 'tab:red')
```
