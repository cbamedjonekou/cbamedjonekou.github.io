```python
# Modules being called
import math as m
import numpy as np
from sympy.solvers import solve
from sympy import plot_implicit, Eq
import sympy as sp
from matplotlib import pyplot as plt
import mpmath as mp
```

```python
def bumFourierSeries():
    Fs = 100
    t = np.linspace(0, 99, 100000)/Fs
    A = 2
    omega_0 = 2 * np.pi * 1
    fundamental = (A/2)
    someK = (2*A/np.pi)
    harmonics = (1/3)*np.sin(3 * omega_0 * t) + (1/5)*np.sin(5 * omega_0 * t) #+ (1/7)*np.sin(7 * omega_0 * t) + (1/9)*np.sin(9 * omega_0 * t) + (1/11)*np.sin(11 * omega_0 * t)
    s = fundamental + (2*A/np.pi) * harmonics
    plt.figure()
    plt.plot(t, s)
```

```python
bumFourierSeries()
```
