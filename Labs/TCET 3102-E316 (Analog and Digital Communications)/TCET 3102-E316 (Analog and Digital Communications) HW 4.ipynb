2. Textbook Page 87 Question 2.4. A periodic waveform is described by f(x) = |sinx| for -π < x < π. By sketching this function, determine if it is an even or odd function, or neither.

```python
# Variables for the plot
x = np.linspace(-np.pi, np.pi)
y = np.abs(np.sin(x))

# The Plot
fig, ax1 = plt.subplots()
ax1.plot(x, y, c= 'r', label= 'Sine Wave')
ax1.set(xlabel= 'Time', ylabel= 'Amplitude', title= 'f(x) = $|\sin(x)|$');
ax1.legend('best')
```

4. Textbook Page 87 Question 2.6. The trigonometric series for a periodic waveform consists of three sine terms consisting of a fundamental, a second harmonic and a third harmonic. The fundamental has amplitude of 1V and provides the reference phrase. The second harmonic has an amplitude of 0.3V and a phase lag of 27° . The third harmonic has an amplitude of 0.5V and a phase lead of 30°. Create the resultant waveform using MATLAB.

```python
# These variables are used to create the fourier series
# Start and Stop indicates my domain for my sampling frequency [samplingfreq; (F_S)], 100 points.
# n1, n2 are the amount of harmonics I want.
start, stop, n1, n2, samplingfreq = 0, 100, 9, 29, 100 

# 't' is for time, and is used to create my 100 Hz time vector
t = np.arange(start, stop, .001) / samplingfreq

# 'A' represents the amplitude 2 volts
A, A2, A3  = 1, 0.3, 0.5

# 'fundamental' is the DC component of the Fourier Series 
fundamental = A/2

# 'signalfreq' is the Signal Frequency (f_0)
signalfreq = 1 

# 'omega' is the Angular Velocity (w_0)
omega = 2 * np.pi * signalfreq

# harmonics1, harmonics2 are AC component of the Fourier series
harmonics1 = (((2*A2)/(np.pi*3) * np.sin((3 * omega * t) - (27*np.pi)/180)) + 
              (((2*A3)/(np.pi*5) * np.sin((5 * omega * t) + np.pi/6))))

# ffs1, ffs2 are the fourier series
ffs1 = fundamental + harmonics1

# Plot

fig, ax1 = plt.subplots()
ax1.plot(t, ffs1)
```
