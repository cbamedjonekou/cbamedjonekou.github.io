3. **Calculate all possible transmitted frequencies for a two-tone modulation system using $21$-MHz carrier w/ $300$-Hz and $700$-Hz modulating signals to represent mark and space. Calculate channel bandwidth required.**

    * This is an amplitude modulation system; therefore, when the carrier is modulated by 300 Hz, the output frequencies will be $21$ MHz and $21$ MHz $\pm$ $300$ Hz. 
    $$\space$$
    * Similarly, when modulated by $700$ Hz, the output frequencies will be $21$ MHz and $21$ MHz $\pm$  700 Hz. Those are all possible outputs for this system. 
    $$\space$$
    * The bandwidth required is therefore $700$ Hz $*$ 2 $= 1400$ Hz, which means that a $1.4$-kHz channel would be adequate.

6. Describe the phase-shift keying (PSK) Process.

    * **Phase-shift keying (PSK)** systems provide a low probability of error. In a PSK system, the transmitted binary data cause the phase of the sine-wave carrier to shift by a defined amount. The modulating signal may cause the carrier to shift to one or more possible phases with respect to its unmodulated (reference) phase.
