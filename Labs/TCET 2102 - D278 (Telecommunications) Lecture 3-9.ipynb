** Filters: Any combination of passive (R, L, and C) and/or active (transistors or operational amplifiers) elements designed to select or reject a band of frequencies is called a filter.**

*  In communication systems, filters are employed to pass those frequencies containing the desired information and to reject the remaining frequencies.

**Filters are employed to filter out any unwanted frequencies, commonly called noise, due to the nonlinear characteristics of some elec- tronic devices or signals picked up from the surrounding medium. In general, there are two classifications of filters:**

* Passive Filters - contains only passive elements (e.g: resistors, capacitors, inductors) **"Passive filters are those filters composed of series or parallel combinations of R, L, and C elements."**

    * Low Pass Filters
    * High Pass Filters
    * Passive Band Filters
    * Stop Band Filters


* Active Filters - contain active elements (i.e.: diodes, Transistors, Operational Amplifiers) in addition to passive elements. **"Active Filters are filters that employ active devices such as transistors and operational amplifiers in combination with R, L, and C elements."**

* ***For filters, we use a normalized plot more often than the $V_O \space vs. \space frequency$ to show the characteristics of filters.***

    * **Normalization is a process where a quantity (e.g.:voltage, current, or impedance) is divided by a quantity of the same unit of measurement to establish a dimensionless level of a specific value (or range).*** A normalized plot in the filter domain can be obtained by dividing the plotted quantity such as $V_O$ with the applied voltage $V_I$ (***Voltage Gain $A_V = \frac{V_O}{V_I}$***) for the frequency range of interest.

* **Since the maximum value of $V_O$ for the low-pass filter of is $V_I$, each level of $V_O$ in the figure below is divided by the level of $V_I$. The result is the plot of $A_V = \frac{V_O}{V_I}$ of the figure below. Note that the maximum value is 1 (Given that $A_V = \frac{V_O}{V_I}; \space and \space V_O = V_I$) and the cutoff frequency is defined at the 0.707 level.***
    

### At any frequency in between the extremes, the output voltage $V_O$ of the figure below can be determined using the voltage divider rule: **



$$V_O = \frac{X_C \angle{-90^{\circ} V_I}}{R-jX_C}$$



* **We can define this relationship in terms of Voltage Gain, $A_V$, as it would be in terms of our *normalized plot* **


$$A_V = \frac{V_O}{ V_I} = \frac{X_C \angle{-90^{\circ}}}{R-jX_C}$$

* **Which can be expanded to the following (by converting the denominator from rectangular to polar form.**

$$A_V = \frac{V_O}{ V_I} = \frac{X_C \angle{-90^{\circ}}}{\sqrt{R^2+X_C^2} \angle{-\tan^{-1}\frac{X_C}{R}}} = \frac{X_C}{\sqrt{R^2+X_C^2}} \angle{-90 + \tan^{-1}\frac{X_C}{R}}$$

* **Further Simplifying the magnitude, we get the following: **

$$A_V = \frac{V_O}{ V_I} = \frac{X_C}{\sqrt{R^2+X_C^2}} = \frac{1}{\sqrt{\left(\frac{R}{X_C}\right)^2 + 1}}$$ 

* **Simplifying the phase angle, we get the following:**

$$ \angle{-90 + \tan^{-1}\frac{X_C}{R}} = \angle{- \tan^{-1}\frac{R}{X_C}}$$

* **We can also define this relationship in terms of frequency by defining $R$ and $X_C$ in terms of frequency, shown below:**

$$A_V = \frac{V_O}{ V_I} =  \frac{1}{\sqrt{\left(\frac{R}{X_C}\right)^2 + 1}} \angle{- \tan^{-1}\frac{R}{X_C}}$$  

* **For $R = \frac{1}{2\pi f C}$ and $X_C = \frac{1}{2\pi f_C C}$**
$$=  \frac{1}{\sqrt{\left(\frac{ \frac{1}{2\pi f C}}{ \frac{1}{2\pi f_C C}}\right)^2 + 1}} \angle{- \tan^{-1}\frac{\frac{1}{2\pi f C}}{ \frac{1}{2\pi f_C C}}}$$  

$$=  \frac{1}{\sqrt{\left(\frac{1}{2\pi f C} * \frac{2\pi f_C C}{1}\right)^2 + 1}} \angle{- \tan^{-1}\frac{1}{2\pi f C} * \frac{2\pi f_C C}{1}}$$ 

$$=  \frac{1}{\sqrt{\left(\frac{f_C}{f}\right)^2 + 1}} \angle{- \tan^{-1}\frac{f_C}{f}}$$ 

$$\textbf{Note that } f_C = \frac{1}{2\pi RC}$$

### For the special frequency at which $X_C = R$ or $f = f_C$ , the magnitude becomes

$$A_V = \frac{V_O}{ V_I} = \frac{1}{\sqrt{\left(\frac{R}{X_C}\right)^2 + 1}} = \frac{1}{\sqrt{\left(\frac{f_C}{f}\right)^2 + 1}} = \frac{1}{\sqrt{2}} \cong 0.707$$

### The impact of the equation $f_C = \frac{1}{2\pi RC}$, extends beyond their relative simplicity. 

* **For any low-pass filter, the application of any frequency less than $f_C$ will result in an output voltage $V_O$ that is at least $70.7\%$ of the maximum.** 

$$\text{For } \space f < f_C, \space V_O \geq 0.707 V_I$$   

* **For any frequency above $f_C$, the output is less than $70.7\%$ of the applied signal.**

$$\text{For } \space f > f_C, \space V_O < 0.707 V_I$$  

* **And at $f_C$, the following is a true observed characteristic.**

$$\mathbf{V_O} \space lags \space \mathbf{V_I} \space by \space 45^{\circ}$$

**A number of methods are used to establish the pass-band characteristic of the Figure above. The one method being discussed here uses both a low-pass and a high-pass filter in cascade, as shown in the Figure below.**

**Having two filters, a low-pass and high-pass cascaded, requires you to realize the following characteristics:**

* **You have two frequencies that pass through your pass-band filter due to the nature of it's two components, the low-pass and high-pass filters**.

    * The low-pass filters, as the name suggest, allows **smaller/lower frequencies to pass** through as they have **outputs (voltage) which are greater than $70.7\%$ of their inputs (voltage),** ($\mathbf{f < f_C; \space V_O > 0.707 V_I \space}$). They block anything higher than the critical frequency ($\mathbf{f = f_C; \space V_O = 0.707 V_I \space}$) from going through, ($\mathbf{f > f_C; \space V_O < 0.707 V_I \space}$).
    <p></p>
    * The high-pass filters, as the name suggest, allows **larger/higher frequencies to pass** through as they have **outputs (voltage) which are greater than $70.7\%$ of their inputs (voltage),** ($\mathbf{f > f_C; \space V_O > 0.707 V_I \space}$). They block anything lower than the critical frequency ($\mathbf{f = f_C; \space V_O = 0.707 V_I \space}$) from going through, ($\mathbf{f < f_C; \space V_O < 0.707 V_I \space}$).
    
    
* **You have two critical/cuttoff frequencies; the area formed between them is called the Bandwidth**.

    *  If the critical frequencies of each filter gets closer and closer together, then there's an increase to the quality factor of the response curve causing the peak values within the pass-band to continue to drop. 


* **You have a frequency, $f_O$, near the center of the pass-band that will pass through both filters with very little degeneration**.

    * For cases where $V_{O(max)} \neq V_{I(max)}$ the bandwidth is defined at $0.707$ of the resulting $V_{O(max)}$.

$$f = f_O; \space BW = 0.707 V_{O(max)} $$  


```python
i(filename='/Users/Chris/Desktop/Images/TCET12.png', width= 400, height= 400)
```

* **Characteristics of pass-band filters rely heavily on design. If there are loadings between stages at each frequency, these loadings will affect the level of $V_O$ at each frequency.**

```python
i(filename='/Users/Chris/Desktop/Images/TCET13.png', width= 400, height= 400)
```

* **Proper design tries to rectify this issue, allowing the level of $V_O$ to be very close to the level of $V_I$ in the pass-band. Unfortunately, it will never equal it exactly.**


* **The pass-band response can be obtained using the series and parallel resonant circuits.**

    * ** However, $V_O$ will not be equal to $V_I$ in the pass-band, but a frequency range in which $V_O$ will be equal to or greater than $0.707*V_{max}$ can be defined.**
    $$\space$$
     
**For the series resonant circuit the following is true:**
    
$$X_L = X_C, \text{ at resonance}$$   
$$V_{O(max)} = \frac{R}{R + R_l} * V_i$$  
$$f_s = \frac{1}{2\pi \sqrt{LC}}$$  
$$Q_s = \frac{X_L}{R + R_l}$$  
$$BW = \frac{f_s}{Q_s}$$

### Converting between series and parallel resonant circuits

**Series to Parallel**

$$R_p = \frac{R_s^2 + X_s^2}{R_s}$$  
$$X_p = \frac{R_s^2 + X_s^2}{X_s}$$  

**Parallel to Series**

$$R_s = \frac{R_p * X_p^2}{R_p^2 + X_p^2} = \frac{R_p}{Q^2 + 1}$$  
$$X_s = \frac{R_p^2 * X_p}{R_p^2 + X_p^2} = \frac{X_p * Q^2}{Q^2 + 1}$$

# Lecture 5: ABCD Parameters (09.25.18)

* Last section we focused on IMPEDANCE (z) Parameters. 

    * **IMPEDANCE (z) Parameters are useful for series connected networks.**
    
    
* This section we'll focus on ABCD Parameters and relate them to IMPEDANCE (z) Parameters.

    * **There is another set of network parameters particularly suited for cascading two-port networks**
  
  
* Unlike in the definition used for $Z$ and $Y$ parameters, notice that $I_2$ is directed away from the port. This is an important point and well discover the reason for it shortly.


```python
i(filename='/Users/Chris/Desktop/Images/TCET22.png', width= 500, height= 300)
```

* **The ABCD matrix is defined as the following:**
$$\space$$ 
$$\Big[\begin{matrix}
  E_1\\
  I_1\\
 \end{matrix}\Big]
 =
\Big[\begin{matrix}
  A & B\\
  C & D\\
 \end{matrix}\Big]
 \Big[\begin{matrix}
  E_2\\
  I_2\\
 \end{matrix}\Big]$$  
$$\space$$  

* **The above can also be written as the following system of equations:**
$$\space$$  
$$E_1 = AE_2 + BI_2$$  
$$I_1 = CE_2 + DI_2$$  
$$\space$$ 

* **We can then find the values of the $ABCD$ Parameters by doing the following:**

    * Taking the first equation of the system you can find Parameters $A$ and $B$.
    $$\space$$  
    * First by setting $I_2 = 0$ and solving for $A$.
    $$\space$$  
    $$E_1 = AV_2 + B(0)$$  
    $$A = \frac{E_1}{E_2}, \text{ for }I_2 = 0$$  
    $$\space$$  
    * Then setting $E_2 = 0$ and solving for $B$.
    $$\space$$  
    $$E_1 = A(0) + BI_2$$  
    $$B = \frac{E_1}{I_2}, \text{ for } E_2 = 0$$  
    $$\space$$
    * Taking the second equation of the system you can find Parameters $C$ and $D$.
    $$\space$$  
    * First by setting $I_2 = 0$ and solving for $C$.
    $$\space$$  
    $$I_1 = CE_2 + D(0)$$  
    $$C = \frac{I_1}{E_2}, \text{ for } I_2 = 0$$  
    $$\space$$  
    * Then setting $E_2 = 0$ and solving for $D$.
    $$\space$$  
    $$I_1 = C(0) + DI_2$$  
    $$D = \frac{I_1}{I_2}, \text{ for } E_2 = 0$$  
    $$\space$$
    
* **As aforementioned above, the usefulness of the ABCD matrix is that cascaded two-port networks can be characterized by simply multiplying their ABCD matrices. This is shown below.**
$$\space$$ 
$$\Big[\begin{matrix}
  E_1\\
  I_1\\
 \end{matrix}\Big]
 =
\Big[\begin{matrix}
  A & B\\
  C & D\\
 \end{matrix}\Big]
 \Big[\begin{matrix}
  E_2\\
  I_2\\
 \end{matrix}\Big]$$  
$$\Big[\begin{matrix}
  E_2'\\
  I_2'\\
 \end{matrix}\Big]
 =
\Big[\begin{matrix}
  A & B\\
  C & D\\
 \end{matrix}\Big]
 \Big[\begin{matrix}
  E_3\\
  I_3\\
 \end{matrix}\Big]$$  
$$\space$$  
    * **Since $\Big[\begin{matrix} E_2\\I_2\\\end{matrix}\Big] = \Big[\begin{matrix}E_2'\\I_2'\\\end{matrix}\Big]$ then the following is true for the cascaded networks:** 
$$\space$$ 
$$\Big[\begin{matrix}
  E_1\\
  I_1\\
 \end{matrix}\Big]
 =
 \Big[\begin{matrix}
  A_1 & B_1\\
  C_1 & D_1\\
 \end{matrix}\Big]
 \Big[\begin{matrix}
  A_2 & B_2\\
  C_2 & D_2\\
 \end{matrix}\Big]
 \Big[\begin{matrix}
  E_3\\
  I_3\\
 \end{matrix}\Big]$$
$$\space$$  
    * **We can the say the following:**
$$\space$$ 
$$\Big[\begin{matrix}
  A_3 & B_3\\
  C_3 & D_3\\
 \end{matrix}\Big]
 =
 \Big[\begin{matrix}
  A_1 & B_1\\
  C_1 & D_1\\
 \end{matrix}\Big]
 \Big[\begin{matrix}
  A_2 & B_2\\
  C_2 & D_2\\
 \end{matrix}\Big]$$
$$\space$$
    * **This simplifying th casceded network formula to the following:**
$$\space$$ 
$$\Big[\begin{matrix}
  E_1\\
  I_1\\
 \end{matrix}\Big]
 =
 \Big[\begin{matrix}
  A_3 & B_3\\
  C_3 & D_3\\
 \end{matrix}\Big]
 \Big[\begin{matrix}
  E_3\\
  I_3\\
 \end{matrix}\Big]$$
$$\space$$

* **In other words, a cascaded connection of two-port networks is equivalent to a single two-port network containing a product of the ABCD matrices.**

**NOTE: It is important to note that the order of matrix multiplication must be the same as the order in which the two ports are arranged in the circuit from signal input to output. Matrix multiplication is not commutative, in general.**

$$[A]*[B] \neq [B]*[A]$$  

#### Relating ABCD Parameters to Z parameters:

* **Recall the following for Z Parameters:**
$$\space$$  
$$z_{11} = \frac{E_1}{I_1}, \text{ when } I_2 = 0$$  
$$z_{12} = \frac{E_1}{I_2}, \text{ when } I_1 = 0$$  
$$z_{21} = \frac{E_2}{I_1}, \text{ when } I_2 = 0$$  
$$z_{22} = \frac{E_2}{I_2}, \text{ when } I_1 = 0$$  
$$\space$$  $$\space$$  
    * **System that also describes them:**
    $$\space$$  
    $$E_1 = z_{11}I_1 + z_{12}I_2$$  
    $$E_2 = z_{21}I_1 + z_{22}I_2$$  
    $$\space$$  
* **Now recall the following for ABCD Parameters:**
$$\space$$  
$$A = \frac{E_1}{E_2}, \text{ for } I_2 = 0$$  
$$B = \frac{E_1}{I_2}, \text{ for } E_2 = 0$$  
$$C = \frac{I_1}{E_2}, \text{ for } I_2 = 0$$  
$$D = \frac{I_1}{I_2}, \text{ for } E_2 = 0$$  
$$\space$$  

* **With a little algebraic manipulation we can get the following**
$$\space$$  
    * **Parameter A (when $I_2 = 0$):**
    $$\space$$  
    $$A = \frac{E_1}{E_2} = z_{11} * \frac{1}{z_{21}} = \frac{z_{11}}{z_{21}}$$  
    $$\text{ Since } z_{11} = \frac{E_1}{I_1} \text{ & } z_{21} = \frac{E_2}{I_1}, \text{ when } I_2 = 0$$  
    $$\space$$  
    * **Parameter B (when $E_2 = 0$):**
    $$\space$$  
        * **Step 1: Use below equations**
        $$\space$$  
        $$E_2 = z_{21}I_1 + z_{22}I_2$$  
        $$E_1 = z_{11}I_1 + z_{12}I_2$$  
        $$\space$$  
        * **Step 2: Set $E_2 = 0$**
        $$\space$$  
        $$0 = z_{21}I_1 + z_{22}I_2$$  
        $$\space$$  
        * **Step 3: Solve for $I_1$**
        $$\space$$  
        $$0 = z_{21}I_1 + z_{22}I_2$$  
        $$z_{21}I_1 = - z_{22}I_2$$  
        $$I_1 = - \frac{z_{22}I_2}{z_{21}}$$  
        $$\space$$  
        * **Step 4: Substitute value for $E_1$ in $B = \frac{E_1}{I_2}$**
        $$\space$$  
        $$B = \frac{E_1}{I_2}$$  
        $$B = \frac{z_{11}I_1 + z_{12}I_2}{I_2}$$  
        $$\space$$                
        * **Step 5: Substitute value for $I_1$ in $B = \frac{z_{11}I_1 + z_{12}I_2}{I_2}$, then simplify.**
        $$\space$$  
        $$B = \frac{z_{11}(- \frac{z_{22}I_2}{z_{21}}) + z_{12}I_2}{I_2}$$  
        $$B = \frac{(- \frac{z_{11}z_{22}I_2}{z_{21}}) + z_{12}I_2}{I_2}$$  
        $$B = \frac{\frac{z_{11}z_{22}I_2}{z_{21}}- z_{12}I_2}{I_2}$$  
        $$B = \frac{I_2(\frac{z_{11}z_{22}}{z_{21}}- z_{12})}{I_2}$$  
        $$B = \frac{z_{11}z_{22}}{z_{21}}- z_{12}$$  
        $$B = \frac{z_{11}z_{22}}{z_{21}}- \frac{z_{12}}{z_{21}}$$  
        $$B = \frac{z_{11}z_{22}- z_{12}z_{21}}{z_{21}}$$  
        $$\space$$  
    * **Parameter C (when $I_2 = 0$):**
    $$\space$$  
    $$z_{21} = \frac{E_2}{I_1}, \text{ when } I_2 = 0$$  
    $$C = \frac{I_1}{E_2}, \text{ for } I_2 = 0$$  
    $$\text{Isolate } E_2: \space z_{21} = \frac{E_2}{I_1} \rightarrow E_2 = z_{21}I_1$$  
    $$\text{Substitute } E_2 \text{ for } z_{21}I_1: \space C = \frac{I_1}{z_{21}I_1}$$  
    $$C = \frac{1}{z_{21}}$$  
    $$\space$$  
    * **Parameter D (when $E_2 = 0$):**
    $$D = \frac{I_1}{I_2}, \text{ for } E_2 = 0$$  
    $$I_1 = - \frac{z_{22}I_2}{z_{21}}$$  
    $$D = \frac{- \frac{z_{22}I_2}{z_{21}}}{I_2}$$  
    $$D = - \frac{z_{22}}{z_{21}}$$

# Lecture 7: Impedance Matching (10.2.18)

* The term impedance matching is rather straightforward; Its simply defined as the process of making one impedance look like another. 

* Frequently, it becomes necessary to match a load impedance to the source or internal impedance of a driving source.

* A wide variety of components and circuits can be used for impedance matching.

* The basis for impedance matching is the Maximum Power-Trasnfer Theorem.
    $$\space$$
    * The maximum power-transfer theorem says that to transfer the maximum amount of power from a source to a load, the load impedance should match the source impedance. **It can be represented by the following equivalencies:**
    $$\space$$
    $$R_{src} = R_L, \text{ for DC sources.}$$
    $$Z_{src} = Z_L, \text{ for AC sources.}$$
    $$\space$$
    * The circuit representing this is shown below:

```python
i(filename='/Users/Chris/Desktop/Images/TCET23.png', width= 500, height= 300)
```

* A key factor of this theorem is that when the load matches the source, the amount of power delivered to the load is the same as the power dissipated in the source. Therefore, transfer of maximum power is only 50% efficient.
$$\space$$  
$$P_I = P_O$$  
$$\space$$  
* The source must be able to dissipate this power. 
* To deliver maximum power to the load, the generator has to develop twice the desired output power.
* As aforementioned, there are a wide variety of components and circuits can be used for impedance matching. **We'll talk about 3 of them:**

    * L-network, which is a simple inductor-capacitor (LC) circuit that can be used to match a wide range of impedances in RF (Radio Frequency) circuits.
    * $\pi$-Network
    * T-Network 

### L-Network

* The primary applications of L-networks involve impedance matching in RF circuits, transmitters, and receivers. 
* L-networks are useful in matching one amplifier output to the input of a following stage. 
* Another use is matching an antenna impedance to a transmitter output or a receiver input. Any RF circuit application covering a narrow frequency range is a candidate for an L-network.

* The low-pass versions are probably the most widely used since they attenuate harmonics, noise, and other undesired signals, as is usually necessary in RF designs.
    *  The key design criteria are the magnitudes and relative sizes of the driving generator output impedance and load impedance.
    

* KEY POINT: The impedances that are being matched determine the Q (quality factor) of the circuit, which cannot be specified or controlled. If it is essential to control Q and bandwidth, a T or -network is a better choice. 


* While the L-network is very versatile, it may not fit every need. There are limits to the range of impedances that it can match. In some instances, the calculated values of inductance or capacitance may be too large or small to be practical for a given frequency range. This problem can sometimes be overcome by switching from a low-pass version to a high-pass version or vice versa.

### $\pi$-Network and T-Network

* The L-network is a real workhorse impedance-matching circuit, however a more complex circuit will provide better performance or better meet desired specifications in some instances.


* That's why we have $\pi$-Networks and T-Networks.
    * The $\pi$-Networks and T-Networks described here will often provide the needed improvement while still matching the load to the source.
    

* The main reason to employ $\pi$-Networks and T-Networks is to get control of the circuit Q (quality factor).
    * Recall L-Networks have a fixed Q. This may not fit your design specifications. 
    * In most cases the Q is very low ($ < 10$). This may be too low for applications where you need to limit the bandwidth to reduce harmonics or help filter out adjacent signals without the use of additional filters. 
    * $\pi$-Networks and T-Networks provide enough variety to fit almost any situation.
