## Brief Theory/Theoretical Background

### S-N NOR Latch

* **SET = 1, RESET = 1** — Tries to set and reset the latch at the same time, 
    * Output is unpredictable, and this inputand condition should not be used.    


* **SET = 0, RESET = 1** — Output will go to Q = 0 and remains there, even after SET returns HIGH.
    * Called resetting the latch.
    
    
* **SET = 1, RESET = 0** — Will produce Q = 1 HIGH and remains there, even after RESET returns HIGH.
    * Called setting the latch. 
    
    
* **NOR gate latch operates similar to the NAND gate latch.**
    * Normal resting state is SET = RESET = 0


### S-N NAND Latch

* **SET = 1, RESET = 1** — Normal resting state, outputs remain in state they were in prior to input.
* **SET = 0, RESET = 1** — Output will go to Q = 1 and remains there, even after SET returns HIGH.
    * Called setting the latch.
    

* **SET = 1, RESET = 0** — Will produce Q = 0 LOW and remains there, even after RESET returns HIGH.
    * Called clearing or resetting the latch. 
    
    
* **SET = 0, RESET = 0**
    * Output is unpredictable, and this input condition should not be used.
