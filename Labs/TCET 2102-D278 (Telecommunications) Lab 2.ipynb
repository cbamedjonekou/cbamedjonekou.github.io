### Theory

**Two-Port Networks**

* This lab focuses on Two-Port Networks (analog telephone networks and function can be describe using them).
* Two port neworks can be either passive or active.
* Many passive telephone networks fall into the Transmission Circuits for telephone communications category.
* Basic Advantage: complex networks can be combined for analysis and synthesis of larger systems.
* We can develop a set of equations (that define the network) using $z$ and $ABCD$ parameters.

**Decibels**

* Decibels are the most used measurement in communications .
* It's a logarithmic expression of ratios (Voltage Gain ($A_V$), Power Gain ($A_P$))
* Several reasons for use are shown below:
    * It's Logrithmic; Great for evaluating and plotting large variations of values w/ equal resolution in each subrange.
    * Makes evaluating insertion loss, gain margin, loop gain, attenuation, and amplification easier
    * Many physical phenomena and human sensation are logarithmic
