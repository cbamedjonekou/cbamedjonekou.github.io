### Three basic logic operations.

* The three basic logic operations are AND, OR, NOT

    * For **AND gates** you only get a true value (or 1, digitally) when both inputs were turned on **(are true)**, and false for everything else

    * For **OR gates**, you only get a true value (or 1, digitally) when either of the imputs or both were turned on **(are true)**, and false for everything else:

    * For **NOT gates**, reverses the values of the input gates.

### NAND, NOR gates

* NAND gates are a combination of AND gates coupled with NOT gates (inverters). Thus the NAND gate produces truth values opposite that of an AND gate. Remember, for AND gates you only get a true value (or 1, digitally) when both inputs were turned "ON", and false for everything else. The opposite is true for NAND gates as aforementioned. For NAND gates you get a true value (or 1, digitally) when the following occurs: True, False; False, True; False, False (two inputs). When both inputs were turned "ON" (True, True) you get a false value for the output.


* NOR gates are a combination of OR gates coupled with NOT gates (inverters). Thus the NOR gate produces truth values opposite that of the OR gate. Remember, for OR gates you only get a true value (or 1, digitally) when either of the imputs or both were turned on, and false for everything else. The opposite is true for NOR gates as aforementioned. For NOR gates, you get a "False" value (or 0, digitally) when either of the imputs or both are turned "ON". You get a "True" output when both inputs are turned "OFF".


* You can build any type of logic circuit using NAND or NOR gates.

### Crux of Lecture 5

* **Boolean Statements are the intermediate step in creating logic circuits. Logic Circuits apply concepts/methods from Propositional Logic (Zero-Order Logic). Propositional Logic makes larger statements (Boolean Statements) from simplistic statements (which can only be *true* or *false*) with propositional connectives ($\cap$, AND; $\cup$, OR; "$\overline{\space \space \space}$", "$\neg$, NOT; "$\oplus$", Exclusive OR (XOR); "$\rightarrow$", IF...THEN; "$\leftrightarrow$", IF AND ONLY IF (IFF)). With the exception of "$\rightarrow$" and "$\leftrightarrow$", the propositional connectives aforementioned represent the logic gates, and logical connectives, used in *crafting truth tables and boolean statements*. The following is the method we'll use to create our own logic circuits.**


* **Method:**

    * Construct Truth Tables
    * From the Truth Tables we created boolean statements
    * Finally we built the logic gates

#### Boolean Theorems

* **Single Variable Theorems**
    $$\space$$
    * AND Gate:
    $$\space$$
        * $X \cap 0 = X * 0 = 0$
        * $X \cap 1 = X * 1 = X$
        * $X \cap X = X * X = X$
        * $X \cap \overline{X} = X * \overline{X} = 0$
    $$\space$$
    * OR Gate:
    $$\space$$
        * $X \cup 0 = X + 0 = X$
        * $X \cup 1 = X + 1 = 1$
        * $X \cup X = X + X = X$
        * $X \cup \overline{X} = X + \overline{X} = 1$
        

* **Multivariable Theorems**
    $$\space$$
    * Communicative Laws:
    $$\space$$
        * $x \cap y = xy = yx = y \cap x$
        * $x \cup y = x + y = y + x = y \cup x$
    $$\space$$
    * Associative Laws:
    $$\space$$
        * $x \cap (y \cap z) = x(yz) = (xy)z = (x \cap y) \cap z = xyz = x \cap y \cap z$
        * $x \cup (y \cup z) = x+(y+z) = (x+y)+z = (x \cup y) \cup z = x+y+z = x \cup y \cup z$
    $$\space$$
    * Distributive Laws:
    $$\space$$
        * $x(y+z) = xy + xz$
        * $(w+x)(y+z) = wy+wz+xy+xz$
        * $x+xy=x$
        * $x+\overline{x}y = x+y$
        * $\overline{x}+xy = \overline{x}+y$
    $$\space$$
    * DeMorgan's Laws:
    $$\space$$
        * Three Variables:
        $$\space$$
            * $\overline{x+y+z} = \overline{x}\space\overline{y}\space\overline{z}$
            * $\overline{xyz} = \overline{x}+\overline{y}+\overline{z}$
        $$\space$$
        * Two Variables:
        $$\space$$
            * $\overline{x+y} = \overline{x}\space\overline{y}$
            * $\overline{xy} = \overline{x}+\overline{y}$
