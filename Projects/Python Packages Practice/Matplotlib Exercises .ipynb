# Matplotlib Exercises 

This problem set focuses on plotting figures using matplotlib. We use the following data.

## Data

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Import matplotlib.pyplot as plt and set %matplotlib inline if you are using the jupyter notebook. 

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

## Exercise 1

* Follow along with these steps:
    * Create a figure object called fig using plt.figure() 
    * Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.
    * Plot (x,y) on that axes and set the labels and titles to match the plot below:

```python
# Creates the figures
fig = plt.figure()

# Adds a set of axes to the figure
ax = fig.add_axes([0,0,1,1])

# Plots on the axes (labels, titles, x vs. y line)
ax.plot(x,y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Title')
```

## Exercise 2

* Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

```python
# Creates the figure
fig = plt.figure()

# Adds the 2 axes to the figure
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.2,.2])
```

* Now plot (x,y) on both axes. And call your figure object to show it.

```python
# Plots on the first axis
ax1.plot(x,y)
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('Title')

# Plots on the second axis
ax2.plot(x,y)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('Title')
```

## Exercise 3

* Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]*

```python
# Creates the figure
fig = plt.figure()

# Adds the 2 axes to the figure
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2,0.5,.4,.4])
```

* Now use x,y, and z arrays to recreate the plot below. Notice the xlimits and y limits on the inserted plot:**

```python
# Plots on the first axis
ax1.plot(x, z, c = 'b')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')

# Plots on the second axis
ax2.plot(x, y, c = 'b')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('zoom')
ax2.set_xlim(20.0, 22,2)
ax2.set_ylim(30, 50)
fig
```

## Exercise 4

* Use plt.subplots(nrows=1, ncols=2) to create the plot below.

```python
# Creates the figure and adds axes based on the values of nrows and ncols
# Unpacks a Tuple
fig, axs = plt.subplots(1, 2)
plt.tight_layout()
```

* Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style

```python
# Plots on axes and shows figure
axs[0].plot(x, y, lw = 3, ls = '--', color = 'b')
axs[1].plot(x, z, lw = 4, color = 'r')
fig
```

* See if you can resize the plot by adding the figsize() argument in plt.subplots() and copying and pasting your previous code.

```python
# Creates the figure and adds axes based on the values of nrows and ncols
# Unpacks a Tuple
fig, axs = plt.subplots(1, 2, figsize = (10,2))
plt.tight_layout()

# Plots on axes and shows figure
axs[0].plot(x, y, lw = 3, color = 'b')
axs[1].plot(x, z, lw = 4, ls = '--', color = 'r')
```
