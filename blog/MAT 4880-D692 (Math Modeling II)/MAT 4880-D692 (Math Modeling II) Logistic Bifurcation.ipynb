# Logist Bifurcation

$x_{n+1} = rx_n(1-x_n)$

$x_0 = 0.35$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
L = [0.35]
r = [2.7, 3.2, 3.9, 2.8]
N = [1000]
for i in range(N[0]):
    x_old = L[-1]
    x_new = r[2]*x_old*(1 - x_old)
    L.append(x_new)
```

```python
L = np.array(L)
for i in range(L.size):
    plt.scatter(i, L[i])
```

```python
r = np.linspace(2.7, 4.0, 500)
N = [1000]

def function_(n, R):
    L = [0.35]
    for i in range(n):
        x_old = L[-1]
        x_new = R*x_old*(1 - x_old)
        L.append(x_new)
    return np.array(L)
```

```python
plt.figure(figsize= (14, 10))

for r_ in r:
    plt.scatter([r_ for i in range(50)], function_(N[0], r_)[-50:],alpha=0.2, s= 1)
```

# Mandelbrot Set

$z_{n+1} = z_n^2 + c$

$c \leftarrow \text{fixed complex number}$

$z_0 = 0$

```python
k = 1000

x, y = np.linspace(-2, 1, 1000), np.linspace(-1, 1, 1000)
xTile, yTile = np.tile(x, (k, 1)), np.tile(y, (k, 1)).transpose()
C, Z, N, M = xTile + yTile*1j, np.full((k,k), 0 + 0*1j), np.full((k,k), True), np.full((k,k), 0)

for i in range(100):
    Z[N] = Z[N]**2 + C[N]
    N[np.abs(Z) > 2] = False
    M[N]+=1

plt.figure(figsize= (14, 10))
plt.imshow(M, aspect= 0.75, interpolation= 'sinc', cmap= 'hot')
```
