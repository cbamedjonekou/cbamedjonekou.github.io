# Section 4.2 - Dynamical Systems

* For this whale problem we are examining two competing species of whales: the Fin Whale, the Blue Whale. We modify the simplest growth model by adding a competition term:

$$\frac{dB}{dt} = r_1B - a_1B^2 - \alpha_1BF$$  
$$\frac{dF}{dt} = r_2F - a_2F^2 - \alpha_2BF$$  

* This model follows the same rules that apply to general dynamical models such as the [Whale Problem 1](https://nbviewer.jupyter.org/github/deaththeberry/ML-AI-HKG_Portfolio/blob/master/Labs/MAT%204880-D692%20%28Math%20Modeling%20II%29/MAT%204880-D692%20%28Math%20Modeling%20II%29%20Whale%20Problem%201.ipynb), so we can still assume the following is true as to keep our model simple:

    * Growth is unrestricted; There's an abundance of room, and other resources need to sustain life. 
    * Therefore, $\text{the growth rate} = rB$ is proportional to the size of the species.
    * There's a self limiting factor; As population of a species increases, individuals of the species compete against each other for resource.
    * Therefore, $\text{the self limiting factor} = aB^2$ inhibits growth.
    * We also have a carrying capacity which can be defined in terms of $r$ and $a$, shown below:
    
    $$\frac{dB}{dt} = rB - aB^2 -\alpha BF= rB\Big(1 - \frac{a}{r}B \Big) -\alpha BF= rB\Big(1 - \frac{B}{K} \Big) -\alpha BF; \text{ where } K = \cfrac{1}{\frac{a}{r}}= \frac{r}{a}$$  
    * **We also have a competition term defined by $\alpha$.**
    * It's reasonable for to assume that for growth models, and for dynamical models in general, growth is linear to smaller populations, and then curtails off for as populations increase. 
        * If $r$, the growth rate, is larger than $a$, the limiting factor, there's more room to grow.
        * The opposite is also true; If $r < a$ there's less room to grow.
    $$\space$$
    * We find equilibrium points (when the system isn't changing) by setting the growth function, or dynamical system, to zero. Shown below:
    
    $$\frac{dB}{dt} = rB\Big(1 - \frac{B}{K} \Big) -\alpha BF= 0$$
    $$\space$$ 
* **As with [Whale Problem 1](https://nbviewer.jupyter.org/github/deaththeberry/ML-AI-HKG_Portfolio/blob/master/Labs/MAT%204880-D692%20%28Math%20Modeling%20II%29/MAT%204880-D692%20%28Math%20Modeling%20II%29%20Whale%20Problem%201.ipynb), formal theory shown below applies:**
    $$\space$$
    * When we model a dynamical model in steady state, we are given the functions:
    
    $$f_1(x_1,...,x_n)$$
    $$.$$
    $$.$$
    $$.$$
    $$f_n(x_1,...,x_n)$$
    $$\space$$
    * which are defined on a subset, called $S$, in the set of real numbers called $R^N$. 
    * The functions $f_1,...,f_n$ represent the rate of change $\Big(\cfrac{\partial x_1}{\partial t},...,\cfrac{\partial x_n}{\partial t}\Big)$ of each variable $x_1,...,x_n$, respectively.
    * A point, $x_1,...,x_n$, in the subset $S$ is called an equilibrium point given that:
    $$\space$$
    $$f_1(x_1,...,x_n) = 0$$
    $$.$$
    $$.$$
    $$.$$
    $$f_n(x_1,...,x_n) = 0$$
    $$\space$$
    * at that point. The rate of change of each of the variables, $x_1,...,x_n$, is equal to zero so the system is at rest.
    * The variables, $x_1,...,x_n$, are called state variables.
    * $S$, the subset of $R^N$, is called the state space.
    * Functions, $f(x_1,...,x_n)$, depend only on the current state, $x_1,...,x_n$, of the system. What this means gives rise to the following:
    $$\space$$
        * Knowledge of the current state of the system is all we need to know to determine the entire future of the system.
        * Once a dynamical system reaches steady state, it remains there forever. This is because the rate of change hits zero. 
            * No change means that you are constant. 
        
 
    

## Solving Whale Problem: Blue vs. Fin Whale Population Model w/ Five-Step Method

### Step 1: List your variables and Assumptions 

**Variables**

$B = \text{number of blue whales}$  
$F = \text{number of finn whales}$  
$g_B = \text{growth rate for blue whales}$  
$g_F = \text{growth rate for finn whales}$  
$c_B = \text{competition term for blue whales}$  
$c_F = \text{competition term for finn whales}$  

**Assumptions**

$g_B = 0.05B\Big(1 - \dfrac{B}{150000}\Big)$  

$g_F = 0.08F\Big(1 - \dfrac{F}{400000}\Big)$  

$c_B = c_F = \alpha BF; \text{ where } \alpha = 10^{-8}$

$\frac{dB}{dt} = g_B - c_B$

$\frac{dF}{dt} = g_F - c_F$

$B, F \geq 0$

$\alpha \text{ is positive and real}$

**Objective**

$\text{Determine if } B\rightarrow 0 \text{ over time}$

### Step 2: Select the modeling approach

* We'll model it as a dynamical system. This dynamical system consists of two state variables: **B, F**.
* The state space is shown below:

$$S = \{(B, \space F):B \geq 0, F \geq 0\}$$

* It also consists of a system of two differential equations.

$$\frac{dB}{dt} = r_1B\Big(1 - \frac{B}{K_1} \Big) -\alpha_1 BF$$  
$$\frac{dF}{dt} = r_2F\Big(1 - \frac{F}{K_2} \Big) -\alpha_2 BF$$  

### Step 3: Formulate the model

* The formulated model is shown below: $r_1 = 0.05; r_2 = 0.08; K_1 = 150000; K_2 = 400000; \alpha_1 = \alpha_2 = 10^{-8}$

$$\cfrac{dB}{dt} = 0.05B\Big(1 - \dfrac{B}{150000}\Big) - 10^{-8}BF$$    
$$\cfrac{dF}{dt} = 0.08F\Big(1 - \dfrac{F}{400000}\Big) - 10^{-8}BF$$  

### Modules

```python
# Modules being called
import math as m
import numpy as np
from sympy.solvers import solve
from sympy import plot_implicit, Eq
import sympy as sp
from matplotlib import pyplot as plt
import pandas as pd
```

### Variables

```python
B,F = sp.symbols('B F')
g_B, g_F = 0.05*B*(1-B/150000), 0.08*F*(1-F/400000)
c_B, c_F = m.pow(10,-8)*B*F, m.pow(10,-8)*B*F
dBdt, dFdt = g_B - c_B, g_F - c_F
```

### Step 4: Solve the Model

* Shown below:

$$\cfrac{dB}{dt} = 0.05B\Big(1 - \dfrac{B}{150000}\Big) - 10^{-8}BF = 0$$    
$$\cfrac{dF}{dt} = 0.08F\Big(1 - \dfrac{F}{400000}\Big) - 10^{-8}BF = 0$$  

### Equilibrium Points

```python
# init_printing() displays LaTeX version of outputs; 'quiet= True' stops 
# init_printing from printing messages regarding its status

sp.init_printing()

# Solution to the system (Equilibrium Points)

solve([dBdt, dFdt], B, F)
```

### Nullclines Solutions

```python
solve(dBdt, B)[1]
```

```python
solve(dFdt, F)[0]
```

```python
solve(dFdt, F)[1]
```

### Functions

```python
def dB_dt(blue, fin):
    """Function that calculates the rate of change for blue whales in the system.
    A differential equation. It takes two inputs."""
    growth_Blue = 0.05*blue*(1-blue/150000)
    competition_Blue = m.pow(10,-8)*blue*fin
    dBxdt = growth_Blue - competition_Blue
    return dBxdt
    
    
def dF_dt(blue, fin):
    growth_Fin = 0.08*fin*(1-fin/400000)
    competition_Fin = m.pow(10,-8)*blue*fin
    dFxdt = growth_Fin - competition_Fin
    return dFxdt
```

### Plotting Vector Field

```python
%matplotlib inline
#----------------------------------------------------------------------------------------
# Creates the superimposed plot for stream plot of the model, as well as dBdt = dFdt = 0.
#----------------------------------------------------------------------------------------

# Part 1: Creates the length of the 'X' and 'Y' Axis 
x, y = np.linspace(0, 200000), np.linspace(0, 600000)
X, Y = np.meshgrid(x, y)

# Part 2: Equations of the nullclines
b, f = x, y
B1, F1, B2, F2 = -0.03*f+150000.0, -0.05*b+400000.0, 0*f, 0*b

# Part 3: The approximated points of the functions dB/dt and dF/dt which we'll use for the plot.
U, V = dB_dt(X, Y), dF_dt(X, Y)

# Part 4: Creating the figure for the plot
fig, ax1 = plt.subplots()

# Part 5: Creating the plots for dB/dt = 0 and dF/dt = 0, the Nullclines
ax1.plot(B1, f, 'r--', lw=2, label='y-nullcline')
ax1.plot(B2, f, 'r--', lw=2)
ax1.plot(b, F1, 'b--', lw=2, label='x-nullcline')
ax1.plot(b, F2, 'b--', lw=2)

# Part 6: Sets the axis, and equilibrium information for the plot
ax1.set(title= 'Whale Problem Model', xlabel= 'Blue Whales', ylabel= 'Fin Whales')
ax1.tick_params(axis='Fin Whales', labelcolor='k')
ax1.legend(loc= 'upper right')

# Part 7: Plots the streamplot which represents the vector plot.
ax1.streamplot(X, Y, U, V)
```

## Sensitivity Analysis

* We test for sensitivity following reason:
    $$\space$$
    * Our $\alpha$ term was given to us for this problem. 
    * We'd like to test how sensitive to change this $"\alpha"$  is.
    * If $\alpha$ is very sensitive a very small change could cause a major change in your result.
    * Likewise, if $\alpha$ is not very sensitive a very small change would do no major change to your result.
    $$\space$$
* And calculate it with the following:
    $$\space$$
    $$\cfrac{\partial x}{\partial y} = \cfrac{x}{y}$$
    $$\space$$

```python
B,F,a = sp.symbols('B F a') # a for \alpha 
g_B, g_F = 0.05*B*(1-B/150000), 0.08*F*(1-F/400000)
c_B, c_F = a*B*F, a*B*F
dBdt, dFdt = g_B - c_B, g_F - c_F
```

```python
# Solution to the system (Equilibrium Points)

solve1 = solve([dBdt, dFdt], B, F)
solve1
```

```python
# solving for \alpha

B_eq = solve1[3][0]
F_eq = solve1[3][1]
```

```python
(sp.diff(B_eq, a) * a)/B_eq
```

```python
(sp.diff(F_eq, a) * a)/F_eq
```

```python
def sense_B(a):
    """Calculates sensitivity of the Blue Whale population in terms of 'alpha'."""
    num1 = 6.66666666666667*m.pow(10,-6)*a
    num2 = 15000000000000.0*m.pow(a,2)-1.0
    subnum1 = (4.5*m.pow(10,18)*a*(8000000.0*a - 1.0))/m.pow((15000000000000.0*m.pow(a,2) - 1.0), 2)
    subnum2 = 1200000000000.0/(15000000000000.0*m.pow(a,2) - 1.0)
    num3 = -subnum1 + subnum2
    denom = 8000000.0*a - 1.0
    return (num1 * num2 * num3)/denom

def sense_F(a):
    """Calculates sensitivity of the Fin Whale population in terms of 'alpha'."""
    num1 = a
    num2 = 15000000000000.0*m.pow(a,2)-1.0
    subnum1 = (30000000000000.0*a*(750000000000.0*a - 400000.0))/m.pow((15000000000000.0*m.pow(a,2) - 1.0), 2)
    subnum2 = 750000000000.0/(15000000000000.0*m.pow(a,2) - 1.0)
    num3 = -subnum1 + subnum2
    denom = 750000000000.0*a - 400000.0
    return (num1 * num2 * num3)/denom
```

**As long as the $\alpha 's$ from below are smaller than our chosen senstivity of $\alpha = 10^{-8}$, then our results aren't sensititve to change in $\alpha$**
